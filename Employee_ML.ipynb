## DATA ENCODING

## 1- One Hot Encoding

```python
# Get Dummies Function

df_getdummies=pd.get_dummies (data=df,columns=["Company","State","Country"])
print(df_getdummies)
```

## Using SK learn 

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
cols_to_encode = ["Company", "State","Country"]
new_encoded_cols_names = []

# Create a list of new columns names
for col in cols_to_encode:
    new_encoded_cols_names += [f"is_{category}" for category in df[col].unique().tolist()]


# Inititalize and fit One Hot Encoder class from sklearn
one_hot_encoder = OneHotEncoder(sparse_output = False,handle_unknown = 'ignore')
encoded_cols = one_hot_encoder.fit_transform(df[cols_to_encode])
print(encoded_cols)

# As the output of the above transformer would be a sparse matrix, we need to convert it back to Pandas dataframe
df_encoded = pd.DataFrame(encoded_cols, columns = new_encoded_cols_names)
print(df_encoded)
df_one_hot_encoded = df.join(df_encoded)


print(df_one_hot_encoded)
```

# Feature Scaling

## 1.Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
## Using Standard Scalar

Columns_scale=df[['Age','Salary']]

# Create a Scalar Object
scaler=StandardScaler()

# Fit the Scalar on the  Data 
scaler.fit(Columns_scale)

#Transforming data
scaled_data=scaler.transform(Columns_scale)

# Converting the scaled data into pandas dataframe

scaled_data=pd.DataFrame(scaled_data,columns=['Age','Salary'])

df_standard=df.copy()

df_standard[['Age','Salary']]=scaled_data

print(df_standard)
```

## 2.Min Max Scalar

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
## Using Min Max Scalar

Columns_scale=df[['Age','Salary']]

# Create a Scalar Object
scaler=MinMaxScaler()

# Fit the Scalar on the  Data 
scaler.fit(Columns_scale)

#Transforming data
scaled_data=scaler.transform(Columns_scale)

# Converting the scaled data into pandas dataframe

scaled_data=pd.DataFrame(scaled_data,columns=['Age','Salary'])

df_minmax=df.copy()

df_minmax[['Age','Salary']]=scaled_data

print(df_minmax)
```
